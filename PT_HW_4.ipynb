Задание 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
    Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
    Если да, найдите ее.

Если принять определения дисперсии для равномерно распределения - D = (b-a)**2/12, то правую границу (b) можно выразить, как b =  sqrt(12*D)+ a

```python
import math
a= 0.5
d= 0.2
b =  math.sqrt(12*d)+ a
print("Левая граница заданного распределения (B)равна:",b)
```

```python
m = (a+b)/2
print("Среднее значение заданого распределения равно:",m)
```

Задание 4. Рост взрослого населения города X имеет нормальное распределение.
    Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
    Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
    а). больше 182 см
    б). больше 190 см
    в). от 166 см до 190 см
    г). от 166 см до 182 см
    д). от 158 см до 190 см
    е). не выше 150 см или не ниже 190 см
    ё). не выше 150 см или не ниже 198 см
    ж). ниже 166 см.


```python
from scipy.stats import norm
```

```python
norm.cdf(0)
```

```python
(182-174)/8
```

```python
norm.cdf((182-174)/8)
```

```python
def rost (a0,a1,d):
    return norm.cdf((a1-a0)/d)
```

```python
rost(174,182,8)
```

```python
print ('a - больше 182 см',(1-rost(174,182,8)))
print ('б - больше 190 см',(1-rost(174,190,8)))
print ('в - от 166 см до 190 см',1-rost(174,166,8)-(1-rost(174,190,8)))
      
print ('г - от 166 см до 182 см',1-rost(174,166,8)-(1-rost(174,182,8)))
print ('д - от 158 см до 190 см',1-rost(174,158,8)-(1-rost(174,190,8)) )
print ('е - не выше 150 см или не ниже 190 см',rost(174,150,8)+(1-rost(174,190,8)) )
print ('ё - не выше 150 см или не ниже 198 см',rost(174,150,8)+(1-rost(174,198,8)))
print ('ж - ниже 166 см.', rost(174,166,8) )
    
```
