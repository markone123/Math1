    2) Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?
    3) Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?


```python
Задание1.Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8.
Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.
```

```python
import math
с100_85 = (math.factorial(100)/(math.factorial(85)*math.factorial(100-85)))
pq = 0.8**85*(1-0.8)**15
p=с100_85 * pq
print('вероятность того, что стрелок попадет в цель ровно 85 раз',p)
```

Задание 2.
Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

```python
p = 0.0004
n = 5000
lambd = p*n
p0 = lambd**0/math.factorial(0)*2.72**(-2)
p2 = lambd**2/math.factorial(2)*2.72**(-2)
print('вероятность, что ни одна из ламп не перегорит в первый день - ', p0)
print('вероятность, что перегорят ровно две - ', p2)
```

Задание 3. 
Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
p = 1/2
q = 1/2
n = 144
k = 70

с144_70 = (math.factorial(144)/(math.factorial(70)*math.factorial(144-70)))
pq2 = p**k*q**(n-k)
p=с144_70 * pq2
print('вероятность, что орел выпадет ровно 70 раз',p)
```

Задание 4.  
В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

```python
p1 = (math.factorial(7)/(math.factorial(2)*math.factorial(7-2)))
p2 = (math.factorial(10)/(math.factorial(2)*math.factorial(10-2)))
p3 = (math.factorial(9)/(math.factorial(2)*math.factorial(9-2)))
p4 = (math.factorial(11)/(math.factorial(2)*math.factorial(11-2)))
print(p1,p2,p3,p4)
p = p1/p2*p3/p4
print('вероятность того, что все мячи белые',p)
p1 = (math.factorial(3)/(math.factorial(2)*math.factorial(3-2)))
p2 = (math.factorial(10)/(math.factorial(2)*math.factorial(10-2)))
p3 = (math.factorial(2)/(math.factorial(2)*math.factorial(2-2)))
p4 = (math.factorial(11)/(math.factorial(2)*math.factorial(11-2)))
print(p1,p2,p3,p4)
p = 1 - p1/p2*p3/p4
print('вероятность того, что хотя бы один мяч белый',p)
```
