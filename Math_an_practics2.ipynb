```python
import math
from decimal import Decimal

def lim (n):
    fact = Decimal(math.factorial(n))
    print('fact ',fact)
    sq = float(math.pow(fact, float(1)/n))
    print('sq ',sq)
    f = n/sq
    print('f ',f)
    return (f)

i=1
xn=float(0)
xnn = float(0)
delta =float(1)
while delta >= 0.0000001 :
    xnn = lim(i)
    delta = float(math.fabs(xn-xnn) )
    print('delta ',delta)
    xn = xnn
    i=i+1
    
```
