Задание 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
z = np.array ([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
print(z)
```

```python
z.shape
```

```python
z.sort()
print(z)
```

```python
np.sum(z)
```

```python
sr_ar = np.sum(z)/z.shape
print('Среднее арифметическое',(np.sum(z)/z.shape))
```

```python
np.std(z)
```

```python
np.std(z,ddof=1)
```

```python
n=20
sigm = 0
i=0
while i < n:
    sigm = sigm + (z[i]-sr_ar)**2
    print(z[i],sigm)
    i=i+1
```

```python
s0 = math.sqrt(sigm/n)
print ('Дисперсия со смещенной оценкой',s0)
print ('Разница с библиотечной функцией',s0-np.std(z))
s1 = math.sqrt(sigm/(n-1))
print ('Дисперсия с несмещенной оценкой',s1)
print ('Разница с библиотечной функцией',s1-np.std(z,ddof=1))
```

```python
#Применим функцию из предыдущего урока

'''
p1 = (math.factorial(7)/(math.factorial(2)*math.factorial(7-2)))
p2 = (math.factorial(10)/(math.factorial(2)*math.factorial(10-2)))
p3 = (math.factorial(9)/(math.factorial(2)*math.factorial(9-2)))
p4 = (math.factorial(11)/(math.factorial(2)*math.factorial(11-2)))
print(p1,p2,p3,p4)
p = p1/p2*p3/p4
print('вероятность того, что все мячи белые',p)
p1 = (math.factorial(3)/(math.factorial(2)*math.factorial(3-2)))
p2 = (math.factorial(10)/(math.factorial(2)*math.factorial(10-2)))
p3 = (math.factorial(2)/(math.factorial(2)*math.factorial(2-2)))
p4 = (math.factorial(11)/(math.factorial(2)*math.factorial(11-2)))
print(p1,p2,p3,p4)
p = 1 - p1/p2*p3/p4
print('вероятность того, что хотя бы один мяч белый',p)
'''


def comb (n, k):
    return int(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))

# Вариант 1 - 1 й ящик - 0 белых мячей, 2-й ящик 3 белых мяча

p1 = comb (5, 0) 
p2 = comb (3, 2)
p3 = comb (8, 2)
p4 = comb (5, 3)
p5 = comb (7, 1)
p6 = comb (12, 4)
print(p1,p2,p3,p4,p5,p6)
p11= (p1*p2)/p3*((p4*p5)/p6)
print('Вероятность по варианту 1 - 1 й ящик - 0 белых мячей, 2-й ящик 3 белых мяча',p11)

# Вариант 2 - 1 й ящик - 1 белый мяч, 2-й ящик 2 белых мяча

p1 = comb (5, 1) 
p2 = comb (3, 1)
p3 = comb (8, 2)
p4 = comb (5, 2)
p5 = comb (7, 2)
p6 = comb (12, 4)
print(p1,p2,p3,p4,p5,p6)
p22= (p1*p2)/p3*((p4*p5)/p6)
print('Вероятность по варианту 2 - 1 й ящик - 1 белый мяч, 2-й ящик 2 белых мяча',p22)

# Вариант 3 - 1 й ящик - 2 белых мяча, 2-й ящик 1 белый мяч

p1 = comb (5, 2) 
p2 = comb (3, 0)
p3 = comb (8, 2)
p4 = comb (5, 1)
p5 = comb (7, 3)
p6 = comb (12, 4)
print(p1,p2,p3,p4,p5,p6)
p33= (p1*p2)/p3*((p4*p5)/p6)
print('Вероятность по варианту 2 - 1 й ящик - 2 белых мяча, 2-й ящик 1 белый мяч',p33)
p = p11+p22+p33

print('Суммарная вероятность того, что 3 мяча белые',p)
```

Задание 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
p1 = 0.9 
p2 = 0.8
p3 = 0.6
pp = (p1+p2+p3)/3
print ('Общая вероятность удачного выстрела', pp)
p11 = p1/pp/3
p22 = p2/pp/3
p33 = p3/pp/3
print ('Вероятность, что выстрел произведен:')
print('Первым спортсменом',p11)
print('Вторым спортсменом',p22)
print('Третьим спортсменом',p33)
```

 Задаие 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
pa = 0.8 
pb = 0.7
pc = 0.9
pp = pa/4+pb/4+pc/4*2
print ('Общая вероятность сдачи сессии', pp)
paa = pa/pp/4
pbb = pb/pp/4
pcc = pc/pp/4*2
print ('Вероятность обучения студента:')
print('На факультете А',paa)
print('На факультете B',pbb)
print('На факультете C',pcc)
```

Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
# Вариант А - выход из строя всех деталей
d1 = 0.1
d2 = 0.2
d3 = 0.25
pa = d1*d2*d3
print('Вероятность выхода из строя всех деталей',pa)
```

```python
pb = (1-d1)*d2*d3+d1*(1-d2)*d3+d1*d2*(1-d3)
print('Вероятность выхода из строя двух деталей',pb)
```

```python
pc = 1-(1-d1)*(1-d2)*(1-d3)
print('Вероятность выхода из строя одной детали',pc)
```

```python
pd = (1-(1-d1)*(1-d2)*(1-d3))- d1*d2*d3
print('Вероятность выхода из строя от одной до двух деталей',pd)
```
