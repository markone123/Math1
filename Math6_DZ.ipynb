1.Задание (теорема сложения)
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

Вероятность находится по формуле P=m/n, где n - число всех равновозможных элементарных исходов эксперимента с подбрасыванием кубика или кости, а m - число тех исходов, которые благоприятствуют событию. Общее число исходов в задаче n=6 Благоприятствуют событию только такие исходы, когда выпадет грань с 2 или 5 очками , таких граней m=2. Тогда искомая вероятность равна P=2/6=1/3=0.33.


2. Задание (теорема умножения)

Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.


Используем результат предыдущей задачи. P1= 0.33, P2 = 0.33 При последовательном подбрасывании, вероятность P=P1 * P2.
P = 0.33 * 0.33 = 0.1089

3 Задание 
Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости.


Сначала разберемся с общим числом исходов эксперимента. когда мы бросали одну кость, все было очевидно,
6 граней - 6 исходов. Здесь костей уже две, поэтому исходы можно представлять как упорядоченные пары чисел вида (x,y),
где x - сколько очков выпало на первой кости (от 1 до 6), y - сколько очков выпало на второй кости (от 1 до 6).
Очевидно, что всего таких пар чисел будет n=6⋅6=36. Нам найти число благоприятствующих событию. Нам нужны два
разных числа (2 и 5), поэтому m=2. Тогда вероятность равна: P=2/36=1/18 = 0,055.

4 Задание  (Геометрическая вероятность +интервалы)


На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?


Представим чертеж

           |-------------------- 15см --------------|
           
А----------D--------------E-------------------------B

|-------- 9 см -----------|


По условию точка С должна лежать на отрезке DE, длина которого равна 9 + 15 - 20 = 4 см

p=|DE|/|AB| = 4/20 = 0,2
