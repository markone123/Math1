```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
plt.gca().set_aspect('equal')
X, Y = np.meshgrid(x,y)
plt.grid(color='lightgray',linestyle='--')
F = X**2 + Y**2 -0.8
plt.contour(X,Y,F,[0])

plt.show()
```

```python
x = np.linspace(-1.0, 1.0, 200)
y = np.linspace(-1.0, 1.0, 200)
plt.gca().set_aspect('equal')
plt.grid(color='lightgray',linestyle='--')
f = x**2 + y**2 -0.8
plt.polar(f)

plt.show()
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

Решение:
Преобразуем 2-е уравнение до вида y = 1 - (1-exp(x))/x


```python
x = np.linspace(-3.0, 3.0, 100)
plt.plot(x, x**2-1)
plt.plot(x, 1 -(1-np.exp(x))/x)
plt.grid(color='lightgray',linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.optimize import fsolve
def eq (p):
    x, y = p
    return(y-x**2+1, np.exp(x)+x*(1-y)-1)
x1, y1 = fsolve(eq, (-2,2))
print(x1,y1)
x2, y2 = fsolve(eq, (2,4))
print(x2,y2)
```
