a) Найти вероятность того, что все карты – крести (вариант 1)
52 карты / 4 масти = 13 карт одной масти, например крести.
13/52 = - вероятость 1 карты
12/51 = - вероятость 2 карты
11/50 = - вероятость 3 карты
10/49 = - вероятость 4 карты

```python
print('вероятость 1 карты', 13/52)
print('вероятость 2 карты', 12/51)
print('вероятость 3 карты', 11/50)
print('вероятость 4 карты', 10/49)
```

```python
print('Вероятность того, что все карты – крести', 13/52*12/51*11/50*10/49)
```

Вариант 2
С(13/4) - число элементарных исходов, благоприятствующих наступлению событий
С(52/4) - Общее число исходов
P = С(13/4)/С(52/4)

```python
import math
p1 = (math.factorial(13)/(math.factorial(4)*math.factorial(9)))
p2 = (math.factorial(52)/(math.factorial(4)*math.factorial(48)))
print('Вероятность того, что все карты – крести',p1/p2)
    
```

```python
б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.
```

 подмножество карт без тузов 52-4 =48


```python
p3 = (math.factorial(48)/(math.factorial(4)*math.factorial(44)))
p2 = (math.factorial(52)/(math.factorial(4)*math.factorial(48)))
#print(p3)
print('вероятность, что среди 4-х карт окажется хотя бы один туз',(p2 - p3)/p2)
```

Задание 3. 
В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
с9_13 = (math.factorial(9)/(math.factorial(3)*math.factorial(6)))
с3_15 = (math.factorial(15)/(math.factorial(3)*math.factorial(12)))
print(с9_13)
print(с3_15)
print('вероятность того, что все извлеченные детали окрашены',с9_13/с3_15)
```

Задание 4.
В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
с100_2 = (math.factorial(100)/(math.factorial(2)*math.factorial(98)))
print(с100_2)
print('вероятность того, что 2 приобретенных билета окажутся выигрышными',1/с100_2)
```
