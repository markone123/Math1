1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
import math

print('Расчет суммы векторов')
x1 = int(input('x1= '))
y1 = int(input('y1= '))
z1 = int(input('z1= '))
x2 = int(input('x2= '))
y2 = int(input('y2= '))
z2 = int(input('z2= '))
x3=x1+x2
y3=y1+y2
z3=z1+z2
print("Сумма векторов равна (%s, %s, %s)" % (x3, y3, z3))
#длина вектора
l=math.sqrt(x3**2+y3**2+z3**2)
print("Длина вектора равна ",l)
```

2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

Прямые не кажутся перпендикулярными. Это связанно из-за различной размерноности осей x и y. Сравните рисунки ниже.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x=np.linspace(-5,5,21)
y=3*x+1
y2=(-1/3)*x+1

plt.plot(x,y)
plt.plot(x,y2)
plt.grid(color='lightgray',linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x=np.linspace(-5,5,21)
y=3*x+1
y2=(-1/3)*x+1
plt.gca().set_aspect('equal')
plt.plot(x,y)
plt.plot(x,y2)

plt.xlabel('x')
plt.ylabel('y')
```

3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
plt.gca().set_aspect('equal')
X, Y = np.meshgrid(x,y)
plt.grid(color='lightgray',linestyle='--')
F = X**2 + Y**2 -0.8
plt.contour(X,Y,F,[0])
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi

u=1.     #x-позиция в центре
v=0.5    #y-позиция в центре
a=2.     #радиус по x-оси
b=1.5    #радиус по y-оси

t = np.linspace(0, 2*pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')
plt.show()
```
